```python
# Mean
import numpy as np

score = [90,80,70,65,110,80,83,67,94]

x = np.mean(score)
print(x)
```

```python
# Median
import numpy as np

score = [90,80,70,65,110,80,83,67,94]

x = np.median(score)
print(x)
```

```python
# Mode
from scipy import stats

score = [90,80,70,65,110,80,83,67,94]

x = stats.mode(score, keepdims=True)
print(x)
```

```python
# Range
import numpy as np

data = [2,3,7,8,9,12,15,20,50,42]

minimum = np.min(data)
maximum = np.max(data)

x = maximum - minimum

print(x)..
```

```python
# Percentile
import numpy as np

data = [90,80,70,65,110,80,83,67,94]

x = np.percentile(data,90)

print(x)
```

```python
#. Quartile dan IQR
import numpy as np

data = [3,4,4,5,5,6,6,7,8,8,9,10]

Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])

IQR = Q3 - Q1

print(Q1, Q3, IQR)
```

```python
# Variance dan Standard Deviation
import numpy as np

data = [86,87,88,86,87,85,86]

mean = np.mean(data)
var = np.var(data)
std = np.std(data)

print(mean, var, std)
```

```python
# 1 Modus dari data pelemparan dadu
from scipy import stats

data = [1,2,3,4,5,5,6,2,3,4,5,6,6,4,3,2,1,4,3,5,6,6,5,4,5]

mode = stats.mode(data, keepdims=True)

print("Modus =", mode)
```

```python
# 2 Persentil ke-25 dan ke-75
import numpy as np

data = [9,10,11,6,8,7,7,5,4,5]

p25 = np.percentile(data,25)
p75 = np.percentile(data,75)

print("Persentil 25 =", p25)
print("Persentil 75 =", p75)
```

```python
# 3 Analisis nilai ujian
import numpy as np

data = [50,54,62,50,52,59,61,63,65,10,53,63,65,50,59,62,50,51,57,60,63,65,65,53,99]

# a. Urutkan data
data_sorted = sorted(data)

# b. Quartile dan IQR
Q1 = np.percentile(data,25)
Q3 = np.percentile(data,75)
IQR = Q3 - Q1

# c. Upper dan Lower Bound
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# d. Outlier
outlier = [x for x in data if x < lower_bound or x > upper_bound]

print("Data terurut =", data_sorted)
print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
print("Lower Bound =", lower_bound)
print("Upper Bound =", upper_bound)
print("Outlier =", outlier)
```

```python
# 4 Standar Deviasi Berat Badan
import numpy as np

# Titik tengah kelas
berat = [43,48,52.5,58,63,68]

# Frekuensi
frekuensi = [2,3,6,4,5,2]

data = []

for i in range(len(berat)):
    data += [berat[i]] * frekuensi[i]

std = np.std(data)

print("Standar Deviasi =", std)
```
